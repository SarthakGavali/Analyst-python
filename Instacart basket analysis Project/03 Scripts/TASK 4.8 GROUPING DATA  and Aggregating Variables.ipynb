```python
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max) #FINDING THE MAXIMUM ORDER PER USER/CUSTOMER
```

```python
#aggregating the customers based on their orders
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
#we shall execute the following command to determine the basic statistics of expenditure grouped by loyalty class
ords_prods_merge.groupby('loyalty_flag').agg({'prices':['mean','min','max']})
```

```python
ords_prods_merge = ords_prods_merge[ords_prods_merge['prices'].isnull()== False]   #Dropping the null rows
```

```python
ords_prods_merge.shape # Determining the number of rows the dataframe now
```

```python
# Creating a spending_flag by classifying spenders 
# creating 'mean_price' column
ords_prods_merge['mean_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.head(15)
```

```python
#creating 'spending flags'
#over or equal the mean price 10 as 'high spender'& under mean price 10 as 'low spender'
ords_prods_merge.loc[ords_prods_merge['mean_price'] >= 10, 'spending_flag'] = 'High spender'
ords_prods_merge.loc[ords_prods_merge['mean_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge.head(15)
```

```python
# checking value counts for 'spending_flag'
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
#step 7 #Creating 'median days since prior order' Column
ords_prods_merge['median_days_since_prior_ord'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge.head(15)
```

```python
#creating a order Frequency flag column 
ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_ord'] > 20, 'ord_frequency_flag'] = 'Non frequent customer'
ords_prods_merge.loc[(ords_prods_merge['median_days_since_prior_ord'] > 10) & (ords_prods_merge['median_days_since_prior_ord'] <= 20), 'ord_frequency_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_ord'] <= 10, 'ord_frequency_flag'] = 'Frequent customer'
```

```python
ords_prods_merge.head(15)
```

```python
# checking value counts for 'frequency flags'
ords_prods_merge['ord_frequency_flag'].value_counts(dropna=False)
```
